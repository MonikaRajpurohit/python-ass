Q4. explain the opertor in python and diffrentiate and provide example how they are used.
Ans: In python there are 7 operator which are used to calculate and sole the problem in programs there are : arithmetic,comparison ,logical,assignment operator .
 ##  Arithmetic Operators

* Arithmetic operators are used to perform mathematical operations.

 Example:
| + | Addition :a = 5; b = 3; print(a + b)
| - | Subtraction : a = 5; b = 3; print(a - b)
| * | Multiplication : a = 5; b = 3; print(a * b)
| / | Division : a = 5; b = 3; print(a / b)
| % | Modulus : a = 5; b = 3; print(a % b)


# Comparison Operators
Comparison operators are used to compare values.

 Example
| == | Equal to : a = 5; b = 5; print(a == b)
| != | Not equal to :a = 5; b = 3; print(a != b)
| > | Greater than : a = 5; b = 3; print(a > b)
| < | Less than : a = 5; b = 3; print(a < b)
| >= | Greater than or equal to : a = 5; b = 5; print(a >= b)
| <= | Less than or equal to :a = 5; b = 5; print(a <= b)

# Logical Operators
Logical operators are used to combine conditional statements.

Example
| and | Logical and : a = 5; b = 3; print(a > 0 and b > 0)
| or | Logical or : a = 5; b = 3; print(a > 0 or b < 0)
| not | Logical not : a = 5; print(not a > 0)

# Assignment Operators
Assignment operators are used to assign values to variables.

 Example
| = | Assign : a = 5
| += | Add and assign :a = 5; a += 3
| -= | Subtract and assign : a = 5; a -= 3
| *= | Multiply and assign : a = 5; a *= 3
| /= | Divide and assign : a = 5; a /= 3
| %= | Modulus and assign : a = 5; a %= 3 .
